```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
file_path = "real_estate_data.csv"  # Ensure this is the correct file path
df = pd.read_csv(file_path)
```

```python
# Function to detect and remove outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Detected {len(outliers)} outliers in '{column}' column.")

    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers from 'price' and 'sqft'
df = remove_outliers(df, 'price')
df = remove_outliers(df, 'sqft')
```

```python
df.price = df.price.fillna(df.price.mean())
df.sqft = df.sqft.fillna(df.sqft.mean())
df.bedrooms = df.bedrooms.fillna(df.bedrooms.mean())
```

```python
x = df[['sqft', 'bedrooms']]
y = df[['price']]
```

```python
# y = a+bx

model = LinearRegression()
model.fit(x, y)
#print(model.predict([[2000, 3]]))
y = model.predict(x)
print(y)
b = model.coef_
a = model.intercept_
```
